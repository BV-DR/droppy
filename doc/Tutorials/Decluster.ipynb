## Data de-clustering

```python
from droppy.TimeDomain import Decluster
import pandas as pd
from datetime import timedelta
```

```python
data = pd.read_csv( "../../TimeDomain/Tests/test_data/hs.csv" , parse_dates=True , index_col=0)
```

```python
data.hs.plot()
```

```python
dec = Decluster( data.hs, threshold = 10.0 , method = "updown",  minSpacing = timedelta(hours = 48) )
```

```python
dec.plot()
```
