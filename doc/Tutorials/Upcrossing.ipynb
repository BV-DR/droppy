## Upcrossing analysis

```python
from droppy.TimeDomain import UpCrossAnalysis
import numpy as np
import pandas as pd
```

```python
fname = "../../TimeDomain/Tests/test_data/eta_surf.csv"
ts = pd.read_csv( fname, index_col = 0 ).iloc[:,0]
ts.plot()
```

```python
up = UpCrossAnalysis.FromTs(  ts , threshold=0.0 , method="upcross" )
```

```python
up.plotTime()
```
